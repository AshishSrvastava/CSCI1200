# 3. Functions & Returns
***

After finishing its task, a function with a return value (sometimes these are called prolific functions) will give back some output value. Let's see how that works!

````python
def my_function(param):
    print("Hello from a function. You have passed the parameter: "+param)
    print("I am returning a word now.")
    ret = "Apple " + param
    return ret

receive = my_function("csci1200")
print("I am outside the function and I have received: "+receive)
````

````python
# Let's write another version of our convert_seconds_to_hours function
# define the function with name "convert_seconds_to_hours" 
# that takes in "seconds" as a parameter
#     # first get minutes from seconds
#     # then get hours from minutes
#     # return number of hours

def convert_seconds_to_hours(seconds):
    mins = seconds/60
    hours = mins/60
    return hours

# define "main" function
#     # welcome the user to the program
#     # ask for the number of seconds to convert to hours
#     # call the "convert_seconds_to_hours" function to calculate hours from seconds
#     # print the number of hours here

def main():
    print("Welcome to the program")
    seconds = float(input("Give a number of seconds to convert to hours:"))
    hours = convert_seconds_to_hours(seconds)
    return hours

# call the "main" function
main()
````

6) (challenge problem) Write a function, `countdown` that __returns__ how many days are between now and an event you are counting down to. This function should have four arguments/parameters: (1) the current year, (2) the absolute day in the current year (as in the question from problem 1 in the pseudocode section, e.g., April 1 is absolute day 91), (3) the year of the event you are counting to, and (4) the absolute day of the event. You may assume that leap years donât exist.

````python
def abs_day(month, day):
        month_name = month.upper()
        months = {"JANUARY":1, "FEBRUARY":2, "MARCH":3, "APRIL":4, "MAY":5, "JUNE":6,
                "JULY":7, "AUGUST":8, "SEPTEMBER":9, "OCTOBER":10, "NOVEMBER":11, "DECEMBER":12}
        month_number = months.get(month_name)
        days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        # NOTE: months has values 1 higher than the respective index in days_in_month, so must subtract 1
        days_until_month = 0
        for i in range(month_number-1):
                days_until_month += days_in_month[i]
        absolute_day = days_until_month + day
        return absolute_day


def countdown(current_year,current_abs_day,event_year,event_abs_day):
        current_day_val = 365 * current_year + current_abs_day
        event_day_val = 365 * event_year + event_abs_day
        return event_day_val - current_day_val
````
