# 1. Writing `if`, `if/elif/else` and `if/else` statements

To write useful programs, we almost always need the ability to check conditions and change the behavior of the program accordingly. **Conditional statements**, sometimes also referred to as **selection statements**, give us this ability. <br>
The simplest form of selection is the `if` statement.  An `if` statement is a control structure that we use to create branches in our code. This means that our code can do one thing under certain circumstances and another thing under different circumstances. The syntax for an `if` statement in python is:

```python
if logical_expression:
    # do something
    print("the logical expression evaluated to true!") # for example
```

In addition to `if`, we also learned how to write these statements with `else` branches in lecture to create more paths for our programs to take.

```python
if logical_expression: # must always begin with "if"
    # do something
    print("first logical expression was true!")
else:  # can have 0 or 1 else
    print("first logical expression was false!")
```


## 1b. Adding in one more piece of the puzzle - `elif`
We can further increase the number of paths at a certain point by adding more branches to our conditionals in the form of one or more `elif`s.


```python
if logical_expression: # must begin with "if"
    # do something
    print("first logical expression was true!")
elif logical_expression2: # can have 0 or more "elif"s
    print("first logical expression was false!")
    print("second logical expression was true!")
else:  # can have 0 or 1 else
    print("first logical expression was false!")
    print("second logical expression was false!")
```

Note that we can use `elif` to create as many paths for our program as needed. For example:

```python
if logical_expression: 
    # do something
    print("first logical expression was true!")
elif logical_expression2: 
    print("first logical expression was false!")
    print("second logical expression was true!")
elif logical_expression3:
    print("first two logical expressions were false!")
    print("third logical expression was true!")
elif logical_expression4:
    print("first three logical expressions were false!")
    print("fourth logical expression was true!")
elif logical_expression5:
    print("first four logical expressions were false!")
    print("fifth logical expression was true!")
else: # Only goes in here if ALL of the logical expressions above were False
    print("Every logical expression above was false!")
```

Be aware that your program will stop evaluating a control structure on the FIRST logical expression it evaluates as true!

# 3. Conditionals and `return`
***
Condinals can be a great way of controlling the flow of functions to ensure the correct value is returned!

````python
# Write a function called "parity" which takes two ints as parameters.
# This function should take its inputs and return:
#      -The string "Both numbers are even" if both integers are even.
#      -The string "Both numbers are odd" if both are odd. 
#      -The string "The numbers have differing parity" otherwise.  

def parity(a,b):
    if a % 2 == 0 and b % 2 == 0:
        print("Both numbers are even!")
    elif a % 2 != 0 and b % 2 != 0:
        print("Both numbers are odd!")
    else:
        print("The numbers have differing parity")


# In main ask for two integers as input. Pass these integers into parity. Print out the result.
def main():
    x = int(input("Enter the first number"))
    y = int(input("Enter the second number"))
    parity(x,y)

main()
````

Write a function `happy_birthday` that takes 4 parameters, the user's birth-month and birth-day and the current month and day. If it is the user's birthday exactly (their birth-month and birth-day are the same as the current month and day), print "Happy birthday!". If it is the user's birth-month (but not day), print "Happy almost birthday!". If it is the user's day of birth (e.g. the 2nd and they were born on the 2nd), but not their birth month (not the same month), print "Happy day of the month of your birthday".

````python
def happy_birthday(birth_month, birth_day, current_month, current_day):
    if birth_month == current_month and birth_day == current_day:
        print("Happy Birthday!")
    elif birth_month == current_month and birth_day != current_day:
        print("Happy almost birthday!")
    elif birth_month != current_month and birth_day == current_day:
        print("Happy day of the month of your birthday")

# happy_birthday("September", 11, "October", 11)
````
