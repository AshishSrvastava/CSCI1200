# 2. Appending and Deleting Elements of Lists
***

Now that we have a better idea of how our loops behave, lets look at how we can append, change, and delete from lists.

Lists are __mutable__. This means that we can change their contents without having a whole new list. This is different than strings. Strings are __immutable__. We cannot change their contents.

```python
# To add an element to a list, we'll use the append function
animals = []
animals.append("mouse")
print(animals)
animals.append("bat")
print(animals)
```

```python
# with strings, we're creating new strings when we add elements to the end
s1 = "ha"
s2 = s1 + "t" # an entirely new string
print(s1)
print(s2)
```

```python
# To change an element in a list, use list[index] = new value
animals = ["mouse", "bat"]
animals[0] = "dog"
print(animals)
animals[1] = animals[1].upper()
print(animals)
```

```python
# we cannot change the value at an index in a string because they are __immutable__
s1 = "hat"
s1[0] = "H"
```

Given a list and a target value, write a function that will return a new list of boolean values, `True` for each element of the list greater than or equal to the target value and `False` for the elements that are not. The original list should be unchanged. What values/types could you call this function with? Give example calls.

```python
def target_values(values_list, target):
    # create empty list to store boolean values
    boolean_list = []
    # while loop counter initialization
    i = 0
    while i < len(values_list):
        # if value at index is greater than target parameter, append True
        if values_list[i] >= target:
            boolean_list.append(True)
        # if value at index is less than target parameter, append False
        else:
            boolean_list.append(False)
        # increment index value
        i += 1
    
    # return list containing boolean values
    return boolean_list

def main():
    # create list holding values
    my_list = [1, 2, 3, 4, 5, 6, 7, 8]
    # define target parameter
    target = 6
    # call target_values to get list of bools in return
    my_bools = target_values(my_list, target)
    # print my_list to verify it wasn't changed
    print("My list: ", my_list)
    # print boolean values to verify
    print("My bools: ", my_bools)
    
# main()
```

Write a function that "mirrors" the word in a given string. For example, given "flower" you should print "flower|rewolf".

```python
def mirror_string(str_in):
    # create empty string to store reversed order
    reverse = ""
    # initialize counter variable to str_in length - 1 (because indexing begins at 0)
    i = len(str_in) - 1
    # iterate through str_in parameter
    while i >= 0:
        # add character from string to reversed string
        reverse += str_in[i]
        # decrement counter
        i -= 1
    # print reversed string
    print(str_in + "|" + reverse)
    
def main():
    # define string to pass to mirror_string function
    word = "flower"
    # call mirror_string function with given parameter
    mirror_string(word)
    
# main()
```

Without using the min or max functions learned in class, write a function that finds the minimum element of a list. Write a function that finds the max element of a list.

```python
def find_min(my_list):
    # initialize while loop condition 
    i = 0
    # set min_value equal to first element in my_list
    min_value = my_list[0]
    # loop through length of my_list
    while i < len(my_list):
        # if element at index is smaller than min_value
        if my_list[i] < min_value:
            # rewrite min_value variable to current value
            min_value = my_list[i]
        # increment
        i += 1
    # return minimum value found in list
    return min_value

def find_max(my_list):
    # initialize while loop condition 
    i = 0
    # set max_value equal to first element in my_list
    max_value = my_list[0]
    # loop through length of my_list
    while i < len(my_list):
        # if element at index is greater than min_value
        if my_list[i] > max_value:
            # rewrite max_value variable to current value
            max_value = my_list[i]
        # increment
        i += 1
    # return minimum value found in list
    return max_value

def main():
    # define list containing int values
    my_list = [3, 2, 5, 6, -1, 5, -7, 10, 12]
    # get minimum element from list
    min_val = find_min(my_list)
    # print min_val to verify
    print(min_val)
    # get maximum element from list
    max_val = find_max(my_list)
    # print max_val to verify
    print(max_val)
    
# main()
```
