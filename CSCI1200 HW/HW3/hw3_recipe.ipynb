__Important__: the only code written outside of a function in this part of your homework should be your call to `main()`. Also, make sure you include comments and function docstrings (see instructions).

```python
def single_ingredient(recipe_multiplier, original_amount, ingredient_name, units):
    """ Prints a string defining the amount, and units of each ingredient necessary

    Parameters:
        recipe_multiplier (float): scaling number for recipe
        original_amount (integer): basis amount for recipe required to make 12 cupcakes
        ingredient_name (string): name of ingredient
        units (string): units of measurement of the ingredient
    Return:
        None
    """
    # Multiply the original amount by the multiplier to get the amount of ingredient
    amount_ingredient = recipe_multiplier * original_amount
    # Round to two decimal places
    rounded_amount = round(amount_ingredient, 2)
    # Print the amount of ingredient with units
    print(str(rounded_amount) + " " + units + " " + ingredient_name )

def grams_to_cups(amount_to_convert, conversion_factor):
    """Returns a conversion for an amount of an ingredient in grams to the same amount in cups

    Parameters:
        amount_to_convert (int): the original amount (in grams) of an ingredient required for 12 cupcakes
        conversion_factor (int): the conversion factor for how many grams of that ingredient are in one cup
    Return:
        amount_cups (float): the amount of the ingredient in cups
    """
    # Divide the amount to convert by the conversion factor to get the converted amount in cups
    amount_cups = amount_to_convert / conversion_factor
    return(amount_cups)


# define a main function
def main():
    # Input number of cupcakes
    str_cupcakes = input("How many cupcakes do you want to make?")
    num_cupcakes = int(str_cupcakes)
    recipe_multiplier = num_cupcakes/12

    # Defining the original amounts of each ingredient:
    original_amount_flour = 100 # grams
    original_amount_sugar = 100 # grams
    original_amount_butter = 100 # grams
    original_amount_eggs = 2 # no units
    original_amount_vanilla_extract = 1 # teaspoons
    original_amount_baking_powder = 1 # teaspoons

    # Print out the relative amount of each ingredient based on the original amount for 12 cupcakes and scaling by the recipe_mulitplier
    single_ingredient(recipe_multiplier, original_amount_flour, "flour", "grams")
    single_ingredient(recipe_multiplier, original_amount_sugar, "sugar", "grams")
    single_ingredient(recipe_multiplier, original_amount_butter, "butter", "grams")
    single_ingredient(recipe_multiplier, original_amount_eggs, "eggs", "")
    single_ingredient(recipe_multiplier, original_amount_vanilla_extract, "vanilla extract", "teaspoon")
    single_ingredient(recipe_multiplier, original_amount_baking_powder, "baking powder", "teaspoon")

    # Converting to cups
    cups_flour = grams_to_cups(original_amount_flour, 120)
    cups_sugar = grams_to_cups(original_amount_sugar, 200)
    cups_butter = grams_to_cups(original_amount_butter, 240)

    # Printing amounts in cups
    print("")
    print("Now in cups!")
    single_ingredient(recipe_multiplier, cups_flour, "flour", "cups")
    single_ingredient(recipe_multiplier, cups_sugar, "sugar", "cups")
    single_ingredient(recipe_multiplier, cups_butter, "butter", "cups")
    single_ingredient(recipe_multiplier, 2, "eggs", "")
    single_ingredient(recipe_multiplier, 1, "vanilla extract", "teaspoon")
    single_ingredient(recipe_multiplier, 1, "baking powder", "teaspoon")

# don't forget to call main() to get your program started, just below this line
main()
```
